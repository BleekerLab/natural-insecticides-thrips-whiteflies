# Data import and transformation
The script imports a table containing:
- the phenotypic classes for whitefly and thrips of the 20 tomato genotypes.
- the stem trichome volatiles normalised values.

```python
import pandas as pd
df = pd.read_csv("../Figure7_PLS-DA/pheno_terpenoids.tsv",sep="\t",index_col=0)
df.head()
```

# Whitefly analysis

## Train and test splits

```python
# get whitefly classes
y = df["wf"].tolist()
```

```python
# get matrix X of volatile values
X = df.iloc[:,2:]
X.head()
```

```python
# Let's split our data into train and test sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

## Initiate a Random Forest model

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=1000)
rf
```

```python
from sklearn.model_selection import cross_val_score

# performs 5 fold cross-validation 
cv_scores = cross_val_score(rf,X,y,cv=5)
cv_scores
```

```python
print("Accuracy: %0.2f (+/- %0.2f)" % (cv_scores.mean(), cv_scores.std() * 2))
```

```python
rf = rf.fit(X_train,y_train)
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

# empty list to collect results

randomForests = []         # will collect rf models
training_accuracies = []   # will collect train_accuracy
test_accuracies = []       # will collect test_accuracy
confusion_matrices = []    # will collect the confusion matrix

for i in range(1,6,1):
    # initialise the RF model
    rf = RandomForestClassifier(n_estimators=1000)
    # splits dataset
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
    # train the model
    rf = rf.fit(X_train,y_train)
    # predict classes for X_test
    y_pred = rf.predict(X_test)
    # calculate train / test accuracies
    train_accuracy = accuracy_score(y_train,rf.predict(X_train))
    test_accuracy = accuracy_score(y_test,y_pred)
    # confusion matrix
    cm = confusion_matrix(y_test,y_pred)
    
    # collect results
    randomForests.append(rf)
    training_accuracies.append(train_accuracy)
    test_accuracies.append(test_accuracy)
    confusion_matrices.append(cm)
```
